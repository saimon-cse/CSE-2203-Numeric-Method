```python
def f(x):
    return (x*x - 4 * x - 10)
```

```python
xValues = []
yValues = []

i = -6.0
while i <= 6.0:
    xValues.append(i)
    yValues.append(f(i))
    i += 0.25 
```

```python
print(xValues)
print(yValues)
```

```python
import matplotlib.pyplot as plt

plt.plot(xValues, yValues)

plt.xlabel("X")
plt.ylabel("f(x)")
plt.title(f"Graph of f(x)")
plt.grid()

plt.axhline(y=0, color = "black") 
plt.axvline(x=0, color = "black")
# plt.xticks(xValues)


plt.show()
```

```python
x1 = -2.0
x2 = 0.0

def bisection(x1, x2):
    for i in range(0, 200):
        
        mid = (x1 + x2) / 2
        if f(mid) == 0:
            # print(f"root: {mid}")
            return mid 
        
        if f(x1) * f(mid) < 0 :
            x2 = mid 
        elif f(mid) * f(x2) < 0:
            x1 = mid
        else: 
            print("No Root Found!")
            break

```

```python
bisection(x1, x2)
```
